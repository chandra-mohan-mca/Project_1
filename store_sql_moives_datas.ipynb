```python
import pandas as pd
```

```python
data=pd.read_csv('imdb_movie_datas_2024.csv')
```

```python
data
```

```python
cursor.execute("""CREATE TABLE imdb (
                                        Movie_Name VARCHAR(100),
                                        Ratings FLOAT,
                                        Voting_Counts INT,
                                        Durations INT,
                                        Genre VARCHAR(100)
                                    );""")
```

```python
# SQL INSERT query
insert_query = """INSERT INTO imdb (Movie_Name, Ratings, Voting_Counts, Durations,Genre) VALUES (%s, %s, %s, %s, %s)"""

# Insert rows from the DataFrame
for index, row in data.iterrows():
      values = (row['Movie_Name'], row['Ratings'], row['Voting_Counts'], row['Durations'],row['Genre'])
      cursor.execute(insert_query,values)

# Commit the transaction
connection.commit()
```
