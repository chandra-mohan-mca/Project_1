```python
import pandas as pd

data=pd.read_csv('c:/Users/vnave/Documents/AI and ML/Scraping data/imdb_movie_datas_2024.csv')
```

```python
data
```

```python
#Top 10 Movies by Rating and Voting Counts
rating_voting_counts=data.nlargest(10, ['Ratings', 'Voting_Counts'])
rating_voting_counts
```

```python
#Voting Trends by Genre
average_voting=data.groupby('Genre')['Voting_Counts'].mean()
average_voting
```

```python
#Genre-Based Rating Leaders
top_rate=data.groupby('Genre')['Ratings'].max()
top_rate
```

```python
#Duration Extremes
shortest_movie=data.loc[data['Durations'].idxmin()]
shortest_movie
```

```python
longest_movie=data.loc[data['Durations'].idxmax()]
longest_movie
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Plot the count of movies for each genre in a bar chart

data['Genre'].value_counts().plot(kind='bar')
plt.Figure(figsize=(10,10))
plt.title('Genre Distribution')
plt.ylabel('Total_Movies')
plt.show()
```

```python
average=data['Durations'].mean().round(2)
average
```

```python
#Show the average movie duration per genre in a horizontal bar chart

average_duration = data.groupby('Genre')['Durations'].mean().sort_values()
plt.barh(average_duration.index, average_duration.values, color='lightgreen')
plt.title('Average Duration by Genre')
plt.xlabel('Average Duration (minutes)')
plt.ylabel('Genre')
plt.show()
```

```python
#Display a histogram or boxplot of movie ratings

plt.boxplot(data['Ratings'])
plt.title('Rating Distribution')
plt.ylabel('Ratings')
plt.show()
```

```python
# Identify genres with the highest total voting counts in a pie chart

sum_vots=data.groupby('Genre')['Voting_Counts'].sum()
plt.pie(sum_vots,labels=sum_vots.index,autopct='%0.2f%%',explode=[0,0,0,0,0,0.3])
plt.title('Most Popular Genres by Voting')
plt.show()
```

```python
sum_vots
```

```python
#Analyze the relationship between ratings and voting counts using a scatter plot

x=data['Ratings']
y=data['Voting_Counts']
plt.scatter(x,y)
plt.Figure(figsize=(10,4))
plt.xlabel('Ratings')
plt.ylabel('Voting_Counts')
plt.title('Relationship between Ratings and Voting Counts')
plt.show()
```

```python
#Use a heatmap to compare average ratings across genres

average_rating=data.groupby('Genre')['Ratings'].mean()
# Convert to DataFrame with a single row or column
average_rating_df = average_rating.to_frame().T 
sns.heatmap(average_rating_df,annot=True, cmap='coolwarm')
plt.title('Ratings by Genre')
plt.show()
```
